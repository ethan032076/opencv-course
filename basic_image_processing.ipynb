```python
cap = cv2.VideoCapture("./data/movie/Cosmos.mp4")
if cap.isOpened() == False:
    sys.exit()
ret, frame = cap.read()
h, w = frame.shape[:2]
fourcc = cv2.VideoWriter_fourcc(*"XVID")
dst = cv2.VideoWriter('output/test.avi', fourcc, 30.0, (w, h))

while True: 
    ret, frame = cap.read()
    if ret == False:
        break

    cv2.imshow('img', frame)
    dst.write(frame)
    if cv2.waitKey(30) == 27:
        break
cv2.destroyAllWindows()
cap.release()
```

```python
gamma = 0.4
img = cv2.imread('./data/src/Berry.jpg')
gamma_dvt = np.zeros((256, 1), dtype=np.uint8)
for i in range(256):
    gamma_dvt[i][0] = 255 * (float(i)/255)**(1.0 / gamma)

img_gamma = cv2.LUT(img, gamma_dvt)
cv2.imshow('img',img)
cv2.imshow('gamma', img_gamma)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def onTrackbar(position):
    global trackValue
    trackValue = position

trackValue = 100
cv2.namedWindow('img')
cv2.createTrackbar('track', 'img', trackValue, 255, onTrackbar)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def print_position(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDBLCLK:
        print(x, y)

img = np.zeros((512, 512), np.uint8)
cv2.namedWindow('img')
cv2.setMouseCallback('img', print_position)
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = np.ones((500, 500, 3)) * 255
cv2.line(img, (0, 0), (150, 190), (255, 0, 0), 2)
cv2.rectangle(img, (100, 25), (300, 150), (0, 255, 0), 4)
cv2.circle(img, (100, 100), 55, (0, 0, 255), -1)
cv2.ellipse(img, (250, 250), (100, 50), 20, 0, 360, (255, 0,  0), 1)
pts = np.array([[100, 30], [200, 30], [200, 80], [100, 50]])
cv2.polylines(img, [pts], False, (100, 255, 0), 3)
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img, "OpenCV", (100, 300), font, 1, (0, 255, 0), 3, cv2.LINE_AA)
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('./data/src/floor.jpg', 0)
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def onTrackbar(position):
    global threshold
    threshold = position

cv2.namedWindow('img')
threshold = 100
cv2.createTrackbar('track', 'img', threshold, 255, onTrackbar)
while True:
    img_th = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 3, threshold)
    cv2.imshow('img', img_th)
    cv2.imshow('src', img)
    if cv2.waitKey(10) == 27: 
        break
cv2.destroyAllWindows()
```
